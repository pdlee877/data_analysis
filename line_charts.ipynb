```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
unrate = pd.read_csv("Documents/Python/unrate.csv")
```

```python
# Generate a line chart that visualizes the unemployment rate from 1948
x_values = unrate["DATE"].head(12)
y_values = unrate["VALUE"].head(12)

plt.plot(x_values, y_values)
# plt.show()
# the x-axis labels are too jumbled
plt.xticks(rotation=90)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Rate Trends, 1948")
plt.show()
```
